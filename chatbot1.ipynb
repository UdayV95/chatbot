```python
from tensorflow import keras
```

```python
 def vectorize_stories(data,word_index=tokenizer.word_index,max_story_len=max_story,max_ques_len=max_question):
    X=[] #stories
    Xq=[] #query/question
    Y=[] #output/correct answer

    for story,query,answer in data:
        x=[word_index[word.lower()] for word in story]
        xq=[word_index[word.lower()] for word in query]
        y=np.zeros(len(word_index)+1)
        y[word_index[answer]]=1
    
        X.append(x)
        Xq.append(xq)
        Y.append(y)
    
    return(pad_sequences(X,maxlen=max_story_len),pad_sequences(Xq,maxlen=max_ques_len),np.array(Y))
```

```python
inputs_train,queries_train,answers_train=vectorize_stories(train_data)
```

```python
inputs_test,queries_test,answers_test=vectorize_stories(test_data)
```

```python
from keras.models import Sequential, Model
from keras.layers import Input, Activation, Dense, Permute, Dropout, add, dot, concatenate, LSTM, Embedding
```

```python
input_sequence=Input((max_story,))
question=Input((max_question,))
```

```python
input_encoder_m=Sequential()
input_encoder_m.add(Embedding(input_dim=vocab_len,output_dim=64))
input_encoder_m.add(Dropout(0.3))
```

```python
input_encoder_c=Sequential()
input_encoder_c.add(Embedding(input_dim=vocab_len,output_dim=max_question))
input_encoder_c.add(Dropout(0.3))
```

```python
question_encoder=Sequential()
question_encoder.add(Embedding(input_dim=vocab_len,output_dim=64, input_length=max_question))
question_encoder.add(Dropout(0.3))
```

```python
input_encoded_m=input_encoder_m(input_sequence)
input_encoded_c=input_encoder_c(input_sequence)
question_encoded=question_encoder(question)
```

```python
match=dot([input_encoded_m,question_encoded],axes=(2,2))
match=Activation('Softmax')(match)
```

```python
response=add([match,input_encoded_c])
response=Permute((2,1))(response)
```

```python
answer=concatenate([response,question_encoded])
```

```python
answer
```

```python
answer=LSTM(32)(answer)
```

```python
answer=Dropout(0.5)(answer)
answer=Dense(vocab_len)(answer)
```

```python
answer=Activation('Softmax')(answer)
```

```python
model=Model([input_sequence, question],answer)
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history=model.fit([inputs_train,queries_train],answers_train,batch_size=32,epochs=150,validation_data=([inputs_test,queries_test],answers_test))
```

```python
import matplotlib.pyplot as plt
print(history.history.keys())
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Model Accuracy")
plt.ylabel("accuracy")
plt.xlabel("epochs")
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.grid()
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
model.save("chatbot_model")
```
